## Code to read the .npz files
For the MedMNIST Classification Decathlon<br>
https://medmnist.github.io/ <br>
The MedMNIST project is described in the short paper https://arxiv.org/abs/2010.14925 <br>

Make a folder for this project in your Google Drive
Download the data files into this folder.


```python
import pandas as pd
import numpy as np
import os
```

```python
# Get a list of the .nps files
list_npz = [x for x in os.listdir() if x[-3::]=='npz']
```

```python
# function to print the data set name and sizes of the train, validation and test sets
def read_npz(npz):
    npz = np.load(npz)
    df= pd.DataFrame.from_dict({item: npz[item] for item in npz.files}, orient='index',dtype='object')
    print(df.index)
    print(df[df.index=='train_images'].values[0][0].shape)
    print(df[df.index=='train_labels'].values[0][0].shape)
    print(df[df.index=='val_images'].values[0][0].shape)
    print(df[df.index=='val_labels'].values[0][0].shape)
    print(df[df.index=='test_images'].values[0][0].shape)
    print(df[df.index=='test_labels'].values[0][0].shape,'\n\n')
```

```python
# report on the contents of all the npsz files
for npz in list_npz:
    print(npz)
    read_npz(npz)
```
